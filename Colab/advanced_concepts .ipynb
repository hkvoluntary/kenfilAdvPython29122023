```python
# conventional function definition:
def sum(x,y):
    return x + y
sum(3,4)
```

```python
# lambda function returns the sum of its two arguments:
sum = lambda x, y : x + y
sum(3,4)
```

```python
#The Python Programming Language offers a filter() function that readily inputs a function & a list as arguments.
#The function returns a new list containing which the functionâ€™s expression evaluates to true.
#  i.e. an item will be produced by the iterator result of filter(function, sequence) if item is included in the sequence "sequence" and if function(item) returns True.
# f = filter(function, sequence)

creature_names = ['Sammy', 'Ashley', 'Jo', 'Olly', 'Jackie', 'Charlie']

def names_vowels(x):
  return x[0].lower() in 'aeiou'

filtered_names = filter(names_vowels, creature_names)

print(list(filtered_names))
```

```python
# With Lamdba
print(list(filter(lambda x: x[0].lower() in 'aeiou', creature_names)))
```

```python
# Map is a Python built-in function that is used to iterate an iterable through a specified function.
#It executes the function through each item in the iterable. It is used to apply a function to a sequence of elements like a list or dictionary.

# map() is a function which takes two arguments:
# r = map(func, seq)

def length(a):
  return len(a)
x  = list(map(length, ("apple","orange","banana")))
print(x)
```

```python
def sum(a,b):
  return a+b
x = list(map(sum,[1,2,3],[1,2,3]))
print(x)
```

```python
def double(value):
  return 2*value
a_list = [1 , 2, 3 , 4, 5 , 6]
map_double_list = list(map(double , a_list))
print(f'Using map function: {map_double_list}')
```

```python
map_lambda_list = list(map(lambda value: 2*value , a_list))
print(f'Multiplying 2 to every element: {map_lambda_list}')
```

```python
def fahrenheit(T):
    return ((float(9)/5)*T + 32)
def celsius(T):
    return (float(5)/9)*(T-32)
temperatures = (36.5, 37, 37.5, 38, 39)
F = map(fahrenheit, temperatures)
C = map(celsius, F)
temperatures_in_Fahrenheit = list(map(fahrenheit, temperatures))
temperatures_in_Celsius = list(map(celsius, temperatures_in_Fahrenheit))
print(f'Fahrenheit: {temperatures_in_Fahrenheit}')
print(f'Celsius: {temperatures_in_Celsius}')


```

```python
C = [39.2, 36.5, 37.3, 38, 37.8]
F = list(map(lambda x: (float(9)/5)*x + 32, C))
print(f'Fahrenheit: {F}')
```

```python
C = list(map(lambda x: (float(5)/9)*(x-32), F))
print(f'Celsius: {C}')
```

```python
numbers = [1, 3, 4, 2]
print(numbers)
print(numbers.sort())  # None
print(numbers)           # [1, 2, 3, 4]

numbers = [1, 3, 4, 2]
print(sorted(numbers)) # [1, 2, 3, 4]
print(numbers)


numbers = [1, 3, 4, 2]
# Sorting list of Integers in descending
numbers.sort(reverse = True)
print(numbers)
```

```python
# A function that returns the 'year' value:
def myFunc(e):
  return e['year']

cars = [
  {'car': 'Ford', 'year': 2005},
  {'car': 'Mitsubishi', 'year': 2000},
  {'car': 'BMW', 'year': 2019},
  {'car': 'VW', 'year': 2011}
]

print(cars)
cars.sort(key=myFunc)
print(cars)
```

```python
# dict
people = {3: "Jim", 2: "Jack", 4: "Jane", 1: "Jill"}
print(people)
print(sorted(people)) # use key to sort
print(people)
print(sorted(people.keys()))
print(sorted(people.values()))
print(people)
print(sorted(people.items()))
```

```python
# sort them by the second name using lambda
name_list = ['Zen Jack', 'Luigi Austin', 'Ben Benson', 'John Ann']
print("Original list", name_list) # Original list ['Zen Jack', 'Luigi Austin', 'Ben Benson', 'John Ann']


name_list.sort(key=lambda name: name.split()[1])
print("Sorted name list", name_list) # Sorted name list ['John Ann', 'Luigi Austin', 'Ben Benson', 'Zen Jack']
```

```python
name_list = ['Zen Jack', 'Luigi Austin', 'Ben Benson', 'John Ann']
print("Original list", name_list) # Original list ['Zen Jack', 'Luigi Austin', 'Ben Benson', 'John Ann']


def sort_by_second_name(name):
    return name.split()[1]


name_list.sort(key=sort_by_second_name)
print("Sorted name list", name_list) # Sorted name list ['John Ann', 'Luigi Austin', 'Ben Benson', 'Zen Jack']
```

```python
try:
   result = 1/0
except ZeroDivisionError as err: # Built-in Python Exceptions - raised when you divide a value or variable with zero.
   print(err)
else:
   print(f"Your answer is {result}")
```

```python
x = [5,8,9,13]

def find_nth_value(x,n):
    try:
        result = x[n]
    except IndexError as err: # Built-in Python Exceptions - occurs when the index of a sequence is out of range
        print(err)
    else:
        print("Your answer is ", result)
```

```python
# Testing
find_nth_value(x,6)
find_nth_value(x,2)
```

```python
def divide(x,y):
    try:
        result = x/y
    except ZeroDivisionError:
        print("Please change 'y' argument to non-zero value")
    except:
        print("Something went wrong")
    else:
        print(f"Your answer is {result}")
    finally: # No matter what the outcome is, it will always run 'finally'
        print("Python programming is fun !")
```

```python
divide(1,0)
divide(3,4)
divide(1,'g')
```
